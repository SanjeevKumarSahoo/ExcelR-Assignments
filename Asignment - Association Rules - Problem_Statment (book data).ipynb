# Apriori Algorithm

# 1) Association rules with 10% Support and 70% confidence

```python
# With 10% Support

frequent_itemsets=apriori(data,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# with 70% confidence

rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions

rules[rules.lift>1]
```

```python
# visualization of obtained rule

plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
plt.figure(figsize=(9,5), dpi= 100)
sns.barplot( data=rules,x="support",y="confidence")
plt.title("Support vs Confidence", fontsize= 18)
plt.xticks(rotation=90)
plt.show()
```

# 2)Association rules with 20% Support and 60% confidence

```python
# With 20% Support

frequent_itemsets2=apriori(data,min_support=0.20,use_colnames=True)
frequent_itemsets2
```

```python
# With 60% confidence

rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.6)
rules2
```

```python
rules2[rules2.lift>1]
```

```python
plt.figure(figsize=(9,5), dpi= 100)
sns.barplot( data=rules2,x="support",y="confidence")
plt.title("Support vs Confidence", fontsize= 18)
plt.xticks(rotation=90)
plt.show()
```

```python
# visualization of obtained rule

plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
